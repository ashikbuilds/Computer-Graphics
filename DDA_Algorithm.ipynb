```python
from matplotlib import pyplot as plt

def DDA(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    steps = max(abs(dx), abs(dy))
    xincrement = dx / steps
    yincrement = dy / steps
    x = x0
    y = y0
    x_coordinates = []
    y_coordinates = []
    for i in range(steps + 1):
        x_coordinates.append(round(x))
        y_coordinates.append(round(y))
        x += xincrement
        y += yincrement
    plt.plot(x_coordinates, y_coordinates, marker='o', color='blue')
    plt.title("DDA Line Drawing Algorithm")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.show()

x0 = int(input("Enter starting x coordinate: "))
y0 = int(input("Enter starting y coordinate: "))
x1 = int(input("Enter ending x coordinate: "))
y1 = int(input("Enter ending y coordinate: "))

DDA(x0, y0, x1, y1)
```
