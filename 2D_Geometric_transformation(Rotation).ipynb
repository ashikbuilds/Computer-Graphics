```python
import matplotlib.pyplot as plt
import math
```

```python
x1, y1 = 5, 10
x2, y2 = 10, 15
x3, y3 = 12, 12
```

```python
angle_rad = math.radians(30)
```

```python
x1_new = x1 * math.cos(angle_rad) - y1 * math.sin(angle_rad)
y1_new = x1 * math.sin(angle_rad) + y1 * math.cos(angle_rad)
x2_new = x2 * math.cos(angle_rad) - y2 * math.sin(angle_rad)
y2_new = x2 * math.sin(angle_rad) + y2 * math.cos(angle_rad)
x3_new = x3 * math.cos(angle_rad) - y3 * math.sin(angle_rad)
y3_new = x3 * math.sin(angle_rad) + y3 * math.cos(angle_rad)

original_x = [x1, x2, x3, x1]
original_y = [y1, y2, y3, y1]

rotated_x = [x1_new, x2_new, x3_new, x1_new]
rotated_y = [y1_new, y2_new, y3_new, y1_new]

plt.plot(original_x, original_y, 'ro-', label='Original Triangle', markersize=3)

plt.plot(rotated_x, rotated_y, 'go-', label='Rotated Triangle', markersize=3)

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')

plt.show()
```
